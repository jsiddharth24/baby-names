```python
header_list = ["Name", "Sex", "Name_Count"]
df = pd.read_csv("C:/Users/Siddharth/Desktop/Blog/Data/names/yob2020.txt", delimiter = ",", names=header_list)
```

```python
df_most_popular_f=df.loc[df['Sex']=='F'].nlargest(50,'Name_Count')
df_most_popular_f.head(10)
```

```python
df['First_Letter']=df['Name'].str[0]
df['Last_Letter']=df['Name'].str[-1]
df['Name_Length']=df['Name'].str.len()
```

```python
df_first_letter_m=df.loc[df['Sex']=='M'].groupby(['Sex','First_Letter']).sum()
df_first_letter_f=df.loc[df['Sex']=='F'].groupby(['Sex','First_Letter']).sum()
df_last_letter_m=df.loc[df['Sex']=='M'].groupby(['Sex','Last_Letter']).sum()
df_last_letter_f=df.loc[df['Sex']=='F'].groupby(['Sex','Last_Letter']).sum()
```

```python
xls_path='C:/Users/Siddharth/Desktop/Blog/Data/names/raw_data.xlsx'

def save_xls(df_list, xls_path):
    with pd.ExcelWriter(xls_path) as writer:
        for n, df in enumerate(df_list):
            df.to_excel(writer,'sheet%s' % n)
        writer.save()
        
save_xls(df_list,xls_path)
```
